```python
def std_per_class(dataset):
    std = torch.zeros(len(dataset.classes))
    image_count = torch.bincount(torch.tensor(dataset.targets))
    for image, label in dataset:
        std[label] = std[label] + torch.std(image)
    return std / image_count
```

```python
import torchvision
import torch
import matplotlib.pyplot as plt
```

```python
transforms = torchvision.transforms.Compose([
    torchvision.transforms.Resize((224, 224)),
    torchvision.transforms.ToTensor()
])
```

```python
path1 = '/content/Sign Language for Numbers'
dataset1 = torchvision.datasets.ImageFolder(path1, transform=transforms)
std_dataset1 = std_per_class(dataset1)
```

```python
class_names = list(dataset1.class_to_idx.keys())
plt.bar(class_names, std_dataset1)
plt.xlabel('Class')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviations per Class')
plt.ylim(0, 1)
plt.show()
```
