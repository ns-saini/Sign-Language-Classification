```python
import os
import glob
import re

# Define the directory where you want to remove files
directory = 'C:\COMP6721\sign_alphabets\dataset5\A - Copy'

# Define the regex pattern for the files you want to remove
regex_pattern = r'depth'  # Example: Remove all .txt files

# Function to recursively search for files in subdirectories
def get_files_recursive(folder_path):
    files = []
    for dirpath, _, filenames in os.walk(folder_path):
        for filename in filenames:
            files.append(os.path.join(dirpath, filename))
    return files

# Get a list of file paths in the directory and its subdirectories
file_paths = get_files_recursive(directory)

# Filter the file paths based on the regex pattern
files_to_remove = [file_path for file_path in file_paths if re.match(regex_pattern, os.path.basename(file_path))]

# Loop through the files to remove and delete them
for file_path in files_to_remove:
    os.remove(file_path)
    print(f"Removed file: {file_path}")
```

```python
import os
import random
import shutil
from torchvision.datasets import ImageFolder
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader

# Define the path to your image folder
data_dir = r"C:\COMP6721\sign_alphabets\dataset5\A"


# Define the path to the folder where you want to copy the random images
target_dir = 'C:\COMP6721\ASL_to_Text_Classification\Test Dataset\Dataset3'

# Create an instance of ImageFolder dataset
dataset = ImageFolder(root=data_dir, transform=ToTensor())

# Create a data loader to load images from the dataset
data_loader = DataLoader(dataset, batch_size=1, shuffle=True)


# Set the number of random images you want to retrieve
num_images = 34

# Get random image paths
random_images = random.sample(dataset.imgs, num_images)

# Extract image paths from tuples
image_paths = [img_path for img_path, _ in random_images]


# Loop through the randomly selected image paths and copy the images to the target folder
for image_path in image_paths:
    image_name = os.path.basename(image_path) # Extract the image file name
    print(image_name)
    target_path = os.path.join(target_dir, image_name) # Create the target path
    print(target_path)
    shutil.copy(image_path, target_path) # Copy the image to the target folder

#     # Print the source and target paths for each copied image
    print(f"Copied {image_path[0]} to {target_path}")
```
